```python
from warnings import WarningMessage
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

from sklearn import datasets
df = pd.read_csv('/content/Kelas.csv')


encode = LabelEncoder()
df['Jurusan'] = encode.fit_transform(df['Jurusan'])
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Jurusan')
y = df['Jurusan']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
```

```python
data_baru = [[0, 88, 77, 95, 60]]
prediksi = knn.predict(data_baru)
print(f'Prediksi target untuk data baru: {prediksi}')
```

# DECISSION TREE

```python
from warnings import WarningMessage
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

from sklearn import datasets
df = pd.read_csv('/content/Kelas.csv')


encode = LabelEncoder()
df['Jurusan'] = encode.fit_transform(df['Jurusan'])
print(df.head())
```

```python

X = df.drop(columns='Jurusan')
y = df['Jurusan']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
tree = DecisionTreeClassifier(random_state=50)
tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = tree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(tree, filled=True,feature_names= )
plt.show()
```
