```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Langkah 2: Membuat dataset sederhana
# Kita akan membuat dataset dengan 2 fitur dan 300 sampel
np.random.seed(42)
X = np.random.rand(300, 2)
# Langkah 3: Menerapkan KMeans
# Kita akan mencoba clustering dengan 3 cluster
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
# Langkah 4: Visualisasi hasil clustering
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200,
alpha=0.75, marker='X')
plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.title('KMeans Clustering')
plt.show()
```

```python
# Langkah 1: Impor library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
# Langkah 2: Membuat dataset sederhana
np.random.seed(42)
X = np.random.rand(300, 2)
# Langkah 3: KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
# Langkah 4: Hierarchical Clustering
linked = linkage(X, 'ward')
hierarchical_labels = fcluster(linked, t=3, criterion='maxclust')
# Langkah 5: Evaluasi KMeans Clustering
kmeans_silhouette = silhouette_score(X, kmeans_labels)
kmeans_davies_bouldin = davies_bouldin_score(X, kmeans_labels)
# Langkah 6: Evaluasi Hierarchical Clustering
hierarchical_silhouette = silhouette_score(X, hierarchical_labels)
hierarchical_davies_bouldin = davies_bouldin_score(X,
hierarchical_labels)
# Langkah 7: Menampilkan hasil evaluasi
print(f'KMeans Clustering:')
print(f'Silhouette Score: {kmeans_silhouette}')
print(f'Davies-Bouldin Index: {kmeans_davies_bouldin}\n')
print(f'Hierarchical Clustering:')
print(f'Silhouette Score: {hierarchical_silhouette}')
print(f'Davies-Bouldin Index: {hierarchical_davies_bouldin}')
# Visualisasi Hasil Clustering
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans Clustering')
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=hierarchical_labels,
cmap='viridis')
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# Langkah 1: Impor perpustakaan yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
# Langkah 2: Membuat dataset sederhana
np.random.seed(42)
X = np.random.rand(300, 2)
# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Langkah 3: Menerapkan DBSCAN
# Menentukan parameter eps dan min_samples
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
# Langkah 4: Evaluasi DBSCAN Clustering
# Menghilangkan label noise (-1) untuk perhitungan silhouette score
labels_without_noise = dbscan_labels[dbscan_labels != -1]
X_without_noise = X_scaled[dbscan_labels != -1]
if len(set(labels_without_noise)) > 1:dbscan_silhouette = silhouette_score(X_without_noise,
labels_without_noise)
else:dbscan_silhouette = -1 # Jika hanya ada satu cluster (selainnoise), silhouette score tidak relevan
dbscan_davies_bouldin = davies_bouldin_score(X_scaled,dbscan_labels)
# Langkah 5: Menampilkan hasil evaluasi
print(f'DBSCAN Clustering:')
print(f'Silhouette Score: {dbscan_silhouette}')
print(f'Davies-Bouldin Index: {dbscan_davies_bouldin}')
# Visualisasi Hasil Clustering
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Load dataset Iris
from sklearn.datasets import load_iris

# Load data & features
iris = load_iris()
X = iris.data

# KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Hierarchical Clustering
linked = linkage(X, 'ward')
hierarchical_labels = fcluster(linked, t=3, criterion='maxclust')

# Evaluation Metrics
kmeans_silhouette = silhouette_score(X, kmeans_labels)
kmeans_davies_bouldin = davies_bouldin_score(X, kmeans_labels)

hierarchical_silhouette = silhouette_score(X, hierarchical_labels)
hierarchical_davies_bouldin = davies_bouldin_score(X, hierarchical_labels)


# Menampilkan hasil evaluasi
print("KMeans Clustering:")
print(f"Silhouette Score: {kmeans_silhouette}")
print(f"Davies-Bouldin Index: {kmeans_davies_bouldin}\n")

print("Hierarchical Clustering:")
print(f"Silhouette Score: {hierarchical_silhouette}")
print(f"Davies-Bouldin Index: {hierarchical_davies_bouldin}\n")

# Visualisasi hasil clustering
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset iris
iris = load_iris()
X = iris.data
y_true = iris.target

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Menerapkan DBSCAN
# Menentukan parameter eps dan min_samples
dbscan = DBSCAN(eps=0.35, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Evaluasi DBSCAN clustering
# Menghilangkan label noise (-1) untuk perhitungan silhouette score
labels_without_noise = dbscan_labels[dbscan_labels != -1]
X_without_noise = X_scaled[dbscan_labels != -1]

if len(set(labels_without_noise)) > 1:
    dbscan_silhouette = silhouette_score(X_without_noise, labels_without_noise)
else:
    dbscan_silhouette = -1  # Jika hanya ada satu cluster (selain noise), silhouette score tidak relevan

dbscan_davies_bouldin = davies_bouldin_score(X_scaled, dbscan_labels)

# Menampilkan hasil evaluasi
print(f'DBSCAN Clustering:')
print(f'Silhouette Score: {dbscan_silhouette}')
print(f'Davies-Bouldin Index: {dbscan_davies_bouldin}')

# Visualisasi hasil clustering
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering Iris')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
